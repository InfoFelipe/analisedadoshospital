### Análise dos dados de Genêro do Hospital Santa Casa - Goiânia - GO

---

#### Importar bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

---

#### Importar a base de dados

```python
baseData = pd.read_csv('agendamentos2.csv')
```

---

```python
dataFrame = pd.DataFrame(baseData)
dataFrame
```

#### Verificando se possui valores nulos na coluna genero

```python
dataFrame.info()
```

#### Total de pessoas cadastradas no ano de 2023

---

```python
dataFrame['data_nascimento'] = pd.to_datetime(dataFrame['data_nascimento'], errors='coerce')
```

```python
dataFrame['ano'] = pd.to_datetime(dataFrame['ano'], errors='coerce')
```

---

#### Total de pessoas do sexo masculino cadastradas no ano de 2023

```python
agendamentoTotalMasc = dataFrame.loc[(dataFrame['ano'].dt.year == 2023) & (dataFrame['genero'] == 1)] 
agendamentoTotalMasc
```

```python
totalAgendamentoMasc = agendamentoTotalMasc.genero.count()
totalAgendamentoMasc
```

---

#### Total de agendamento do genêro Feminino

```python
agendamentoTotalFeme = dataFrame.loc[(dataFrame['ano'].dt.year == 2023) & (dataFrame['genero'] == 2)] 
agendamentoTotalFeme
```

```python
totalAgendamentoFeme = agendamentoTotalFeme.genero.count()
totalAgendamentoFeme
```

---

#### Total de agendamento do genêro Neutro

```python
agendamentoTotalNeutro = dataFrame.loc[(dataFrame['ano'].dt.year == 2023) & (dataFrame['genero'] == 3)] 
agendamentoTotalNeutro
```

```python
totalAgendamentoNeutro = agendamentoTotalNeutro.genero.count()
totalAgendamentoNeutro
```

---

#### Apresentar graficamente os valores totais dos dados masculino, feminino e neutro

```python
totalAgendamentoAnual = [totalAgendamentoMasc, totalAgendamentoFeme, totalAgendamentoNeutro]
totalAgendamentoAnual
```

```python
sexo = ("Masculino", "Feminino", "Neutro")
```

```python
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')
barplot = plt.bar(sexo, totalAgendamentoAnual, color='blue')
plt.xticks(sexo)
plt.ylabel('Agendamentos - 2023', fontsize=12)
# plt.xlabel('Sexo', fontsize=12)
plt.ylim([0, 12000])
plt.bar_label(barplot, labels = totalAgendamentoAnual, label_type ='edge')
plt.tick_params(axis='x', labelsize=12, color='black')
plt.tick_params(axis='y', labelsize=12, color='black')
plt.title('Total de agendamentos por gênero Janeiro até Maio - 2023')
# plt.figure(figsize=(10, 6))
# plt.style.use('ggplot')
plt.savefig("total.pdf")
plt.show()
```
