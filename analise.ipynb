```python
meses = ("Janeiro", "Fevereiro", "Março", "Abril")
quantidadeagendamentos = {
    'Aguardando': (quantidadeAguardandoJaneiro, quantidadeAguardandoFevereiro, quantidadeAguardandoMarco, quantidadeAguardandoAbril),
    'Confirmado': (quantidadeConfirmadoJaneiro, quantidadeConfirmadoFevereiro, quantidadeConfirmadoMarco, quantidadeConfirmadoAbril),
    'Cancelado': (quantidadeCanceladoJaneiro, quantidadeCanceladoFevereiro, quantidadeCanceladoMarco, quantidadeCanceladoAbril),
    'Não Registrado': (quantidadeNaoCadastradoJaneiro, quantidadeNaoRegistradoFevereiro, quantidadeNaoregistradoMarco, quantidadenaoRegistradoAbril
),
}


label = np.arange(len(meses))
width = 0.18
multiplier = 0
font = 20

fig, ax = plt.subplots(layout='constrained', figsize = (17, 10))

for atribute, measurement in quantidadeagendamentos.items():
    offset = width * multiplier
    rects = ax.bar(label + offset, measurement, width, label = atribute)
    ax.bar_label(rects, padding=10)
    multiplier += 1

ax.set_ylabel('Agendamentos', fontsize=15, color='blue')
ax.set_title('Quantidade de agendamento por status Janeiro até Abril - 2023', fontsize = font)
ax.set_xticks(label, meses, color='blue', fontsize=15)
ax.legend(fontsize = 12)
ax.set_ylim(0, 12000)
ax.tick_params(axis='x', labelsize=14, color='black')
ax.tick_params(axis='y', labelsize=14, color='black')
# plt.savefig("analise1.pdf")
plt.show()
```
